## How to Use the Exercise File

You have two main options for completing the exercises.

### Option 1: Run the Script from the Terminal (Recommended)

You can run the entire script directly from your terminal. This will execute all exercises in order.

1.  Open a terminal or command prompt.
2.  Make sure your virtual environment is activated.
3.  Run the following command:
    ```bash
    python week1/week1_day2_exercises.py
    ```
4.  The script will print the output for each exercise. You can open the `.py` file in a text editor to see the code and follow along.

### Option 2: Copy-Paste into a New Notebook

If you prefer the notebook environment, you can copy and paste the code for each exercise into a new, blank notebook.

1.  Create a new, empty Jupyter notebook.
2.  Open the `week1/week1_day2_exercises.py` file in a text editor.
3.  Copy the code for the setup and each exercise into separate cells in your new notebook.
4.  Run the cells one by one.

This approach should prevent any file corruption issues while still allowing you to work in a notebook.
